```python
def find_sheet_name(file_path, possible_names):
    xls = pd.ExcelFile(file_path)
    for name in xls.sheet_names:
        if all(k in name for k in possible_names):
            return name
    return None
```

```python
def find_start_row(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    for i, row in df.iterrows():
        row_str = ' '.join(str(cell).lower() for cell in row)
        if (
                'primary' in row_str and
                'diagnosis' in row_str and
                'emergency' in row_str
        ):
            return i
    return None
```
